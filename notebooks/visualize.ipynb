# Visualization

This notebook is dedicated to the visualization of the different layers of activations values obtained from the LLM and different datasets. 

```python
from sklearn.decomposition import PCA

def get_pca(data,n_components,data_fit=None):
    pca = PCA(n_components=n_components)
    if data_fit is None:
        return pca.fit_transform(data)
    pca.fit(data_fit)
    return pca.transform(data)
```

## Cities dataset

First, let's visualize the principal components of the activations values obtained from the LLM and the cities dataset.

```python
import pandas as pd
cities_act_5 = pd.read_csv('../activations/cities/layer_5_output_cities.csv')
cities_act_10 = pd.read_csv('../activations/cities/layer_10_output_cities.csv')
cities_act_18 = pd.read_csv('../activations/cities/layer_18_output_cities.csv')
cities_act_25 = pd.read_csv('../activations/cities/layer_25_output_cities.csv')
cities_act_35 = pd.read_csv('../activations/cities/layer_35_output_cities.csv')
```

```python
X_pca = [get_pca(cities_act_5, 2), get_pca(cities_act_10, 2), get_pca(cities_act_18, 2), get_pca(cities_act_25, 2), get_pca(cities_act_35, 2)]
```

```python
#plot the pca_result with the label as color
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
for i in range(5):
    plt.subplot(2,3,i+1)
    plt.scatter(X_pca[i][:,0],X_pca[i][:,1],c=cities_act_5.iloc[:, -1],cmap='bwr')
    plt.title('Layer ' + str(i*5+5))
    plt.xlabel('PC1')
    plt.ylabel('PC2')
# Add a legend for 0 (blue) and 1 (true) label in the plot

plt.figlegend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='b', markersize=10, label='0'),  
                        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='r', markersize=10, label='1')], 
                 labels=['0', '1'], bbox_to_anchor=(0.7, 0.38), loc='lower right', borderaxespad=0.5, title='Label')
plt.suptitle('Cities dataset visualizations with PCA', fontsize=16)
plt.show()
    
```
